### 74. Search a 2D Matrix
You are given an m x n integer matrix matrix with the following two properties:

Each row is sorted in non-decreasing order.
The first integer of each row is greater than the last integer of the previous row.
Given an integer target, return true if target is in matrix or false otherwise.

You must write a solution in O(log(m * n)) time complexity.

```python
from typing import List
class twodMatrix:
    def __init__(self, matrix:List[List[int]], target:int) -> bool:

        # general test case if the matrix is empty or it does contain a row with out column 
        if not matrix or not matrix[0]:
            return False
        
        row = len(matrix) #fetches the rows of the matrix 
        cols = len(matrix[0]) #fetches the columns of the matrix 

        
        ## To speed up the time complexity process , lets go with the binary search for the rows and the binary search for the columns 

        # binary search for the rows 
        start_row_index = 0 
        end_row_index = row-1 

        while start_row_index <= end_row_index: ##as long as this condition passes 

            mid = start_row_index + end_row_index // 2  ##now that we set the mid index to traverse between 

            # now we need to binary search the exact row list that has the range the target falls 
            if target < matrix[mid][0]:
                start_row_index = mid - 1 
            
            elif target > matrix[mid][cols-1]:
                end_row_index = mid + 1 
            
            else:   #if none of the start or the end row has the column index element 
                row_final = mid    # then the mid row has the range of the elements 
                break 

        else:
            return False 


        ## now that we found the row via Binary search 

        left, right  = 0 , cols-1 

        while left <= right:
            mid = left+right // 2 # now that we set the mid index to traverse between front and back 

            if matrix[row_final][mid] == target:
                return True 
            elif matrix[row_final][mid] < target:
                left = mid+1 
            else:
                right = mid-1 
            
        else:
            return False 
             
                
            
            



```
